```python
import pandas as pd
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt

df=pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")
```

```python
df
```

```python
#Oppgave 7.1 1a

df2 = df[["Year", "log p", "log q", "log h"]].copy()

df2["price"] = np.exp(df2["log p"])
df2["crop"] = np.exp(df2["log q"])
df2["harvest"] = np.exp(df2["log h"])
df2
```

```python
#Oppgave 7.1 1b

fig, axs = plt.subplots(1,2, figsize=(15,5))

axs[0].set(ylim=(0,500))
axs[1].set(ylim=(40,90))
axs[0].set_xlabel('Year', loc='center')
axs[1].set_xlabel('Year', loc='center')
axs[0].set_ylabel('Price', loc='center')
axs[1].set_ylabel('Quantity', loc='center')

axs[0].plot(df2["Year"],df2["price"], label = "Price", marker = "o", color="black", markerfacecolor="white", linewidth=1)
axs[1].plot(df2["Year"],df2["crop"], label = "Crop", linestyle='dotted', marker = ".", color="black", markerfacecolor="white")
axs[1].plot(df2["Year"],df2["harvest"], label = "Harvest", linestyle='dashed', marker = ".", color="black")

axs[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.12),
          fancybox=False, shadow=False, frameon=False)
axs[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.12),
          fancybox=False, shadow=False, ncol=2, frameon=False)

axs[0].set_title("Price p (dollars per thousand)")
axs[1].set_title("Crop q and harvest h (millions)")
```

```python
#Oppgave 7.1 2abc

q = range(20, 105, 5)

df3 = pd.DataFrame(q, columns=['Q'])
df3["log Q"] = np.log(df3["Q"])
df3["Supply (log P)"] = -2.0+1.7*np.log(df3["Q"])
df3["Demand (log P)"] = 8.5-0.82*np.log(df3["Q"])
df3["Supply (P)"] = np.exp(df3["Supply (log P)"])
df3["Demand (P)"] = np.exp(df3["Demand (log P)"])
df3
```

```python
#Oppgave 7.1 2d

fig, ax = plt.subplots()

ax.plot(df3["Q"], df3["Supply (P)"], label="Supply")
ax.plot(df3["Q"], df3["Demand (P)"], label="Demand")
ax.set_xlabel("Quantity (Q)", loc="center")
ax.set_ylabel("Price (P)", loc="center")
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), fancybox=False, shadow=False, ncol=2, frameon=False)
ax.spines["top"].set_color("none")
ax.spines["right"].set_color("none")
```

```python
#Oppgave 7.1 3a

df3["New supply (log P)"] = -2.0+1.7*np.log(df3["Q"])+0.4
df3["New supply (P)"] = np.exp(df3["New supply (log P)"])
df3
```

```python
#Oppgave 7.1 3b

fig, ax = plt.subplots()

ax.plot(df3["Q"], df3["Supply (P)"], label="Supply")
ax.plot(df3["Q"], df3["Demand (P)"], label="Demand")
ax.plot(df3["Q"], df3["New supply (P)"], label="New supply", linestyle="dashed")
ax.set_xlabel("Quantity (Q)", loc="center")
ax.set_ylabel("Price (P)", loc="center")
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), fancybox=False, shadow=False, ncol=3, frameon=False)
ax.spines["top"].set_color("none")
ax.spines["right"].set_color("none")
```

```python
#Oppgave 7.2 1a

q, p = sp.symbols("q p", real=True, positive=True)

supply = sp.Eq(sp.log(p),-2.0+1.7*sp.log(q))
supply_solved = sp.solve(supply,sp.log(q))

#Oppgave 7.2 1b
demand = sp.Eq(sp.log(p),8.5-0.82*sp.log(q))
demand_solved = sp.solve(demand,sp.log(q))


print(f"{supply_solved} Priselastisiteten er 0.58")
print(f"{demand_solved} Priselastisiteten er -1.2")
```
